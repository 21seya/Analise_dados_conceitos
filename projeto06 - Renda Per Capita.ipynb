```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Lendo o arquivo 
df_renda = pd.read_excel('Dados_Pib.xlsx')

df_renda.head()
```

```python
df_renda.groupby(by=['Territorialidades','Ano']).mean()
```

```python
# Sistema de Grids

# Cor Fundo de Fundo
Cor_Fundo = '#f5f5f5'

Grid_Graficos = sns.FacetGrid(df_renda,col='Territorialidades',hue='Territorialidades',col_wrap=4)

#Adicionar gráfico linhas de cada g´rafico disponível
Grid_Graficos = Grid_Graficos.map(plt.plot,'Ano','PIB per capita')

#Adicionar uma sombra
Grid_Graficos = Grid_Graficos.map(plt.fill_between,'Ano','PIB per capita',alpha=0.2).set_titles('{col_name} Territorialidades')

#Filtro de Titulo
Grid_Graficos = Grid_Graficos.set_titles('{col_name}')


#Adicionar subtitulo 
Grid_Graficos = Grid_Graficos.figure.suptitle(
    'Evolução da Renda per capita por Estado '
)

# Ajustando
plt.subplots_adjust( top=0.92 )


```
