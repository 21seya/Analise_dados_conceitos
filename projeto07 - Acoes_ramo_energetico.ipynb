```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo a base de dados
df_energia = pd.read_excel('Dados_Empresas_Energia.xlsx')

# Verificando
df_energia.head()
```

```python
#Gr√°fico 

#Estilo do gr√°fico 

plt.style.use('seaborn-darkgrid')

#Tamanho 
plt.figure(figsize=(16,7))

#Titulo 
plt.title('Analise de A√ß√µes de Empresas no Segmento de Energia',loc='left',fontsize=18, fontweight=0 )

#Plot da Petrobras 

plt.plot(df_energia.index,df_energia['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7)

#Texto da Petrobr√°s 
plt.text(df_energia.index[-1],df_energia['Petrobras'].tail(1),'Petrobras', color='#008c4a', size='large', horizontalalignment='left')


#plot todas as colunas 
for Coluna in df_energia.columns[1:]:
    #Plot ouras a√ß√µes
    plt.plot(df_energia.index,df_energia[Coluna], color='gray', linewidth=2, alpha=0.7)

    #Texto das outras A√ß√µes
    plt.text(df_energia.index[-1],df_energia[Coluna].tail(1),Coluna, color='gray', size='large', horizontalalignment='left')


#Labels 
plt.xlabel('Periodo')
plt.ylabel('Pre√ßo de Fechamento:(R$)')










```
